```python
import json
import sys 
sys.path.append('/home/nbuser/library/')


import pandas as pd
import requests
```

```python
def getpop(lat,lon):
    url = 'https://koordinates.com/services/query/v1/vector.json?'

    params =  {
        'key': '726d9dfe93304d728a0316c9f661f0f8',
        'layer': '104612',
        'x': lon,
        'y': lat
    }
    response = requests.get(url,params=params)
    if response.status_code != 200:
        return pd.Series({'Population': response.status_code})
    curpop = response.json()['vectorQuery']['layers']['104612']['features'][0]['properties']['C13_CURPop']
    return pd.Series({'Population': curpop})
    
```

```python
df = pd.read_csv("Properties.csv")
```

```python
df['Population']=df.apply(lambda x:getpop(x['Latitude'],x['Longitude']),axis=1)
```
